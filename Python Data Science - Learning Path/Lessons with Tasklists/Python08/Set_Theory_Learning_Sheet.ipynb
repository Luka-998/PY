## 🔷 Introduction to Set Theory
Set theory is the branch of mathematical logic that studies sets, which are collections of objects. It provides a fundamental framework for mathematics.

### **Basic Definitions**  
A **set** is a well-defined collection of distinct objects, called **elements**.

- **Notation:** A set is usually denoted by a capital letter (e.g., $A, B, C$), and elements are written within curly braces:  
  $$ A = \{1, 2, 3, 4, 5\} $$
- **Membership:** If an element $x$ belongs to a set $A$, we write $x \in A$, otherwise $x \notin A$.

## ✨ Types of Sets

| Type | Definition | Example |
|------|------------|---------|
| **Finite Set** | A set with a countable number of elements. | $ A = \{2, 4, 6, 8\} $ |
| **Infinite Set** | A set with an uncountable number of elements. | $ B = \{1, 2, 3, \dots\} $ (Natural numbers) |
| **Empty (Null) Set** | A set with no elements, denoted as $\emptyset$ or $\{\}$. | $ C = \{\} $ |
| **Singleton Set** | A set with only one element. | $ D = \{7\} $ |
| **Equal Sets** | Two sets having exactly the same elements. | $ X = \{a, b, c\}, Y = \{c, a, b\} $ |
| **Subset ($\subseteq$)** | If all elements of set $A$ are in set $B$, then $A \subseteq B$. | $ A = \{1,2\}, B = \{1,2,3\} \Rightarrow A \subseteq B $ |
| **Proper Subset ($\subset$)** | $A$ is a subset of $B$, but $A \neq B$. | $ A = \{1,2\}, B = \{1,2,3\} $ |
| **Universal Set ($U$)** | The set containing all elements under consideration. | $ U = \{1,2,3,4,5,6,7,8,9\} $ |

## 📚 Set Operations

### **Union ($\cup$)**
The union of two sets $A$ and $B$ is the set of all elements in either $A$ or $B$.

$$ A \cup B = \{x \mid x \in A \text{ or } x \in B\} $$

**Example:**

$$ A = \{1, 2, 3\}, B = \{3, 4, 5\} \Rightarrow A \cup B = \{1, 2, 3, 4, 5\} $$

---

### **Intersection ($\cap$)**
The intersection of $A$ and $B$ consists of elements that are in both sets.

$$ A \cap B = \{x \mid x \in A \text{ and } x \in B\} $$

**Example:**

$$ A = \{1, 2, 3\}, B = \{3, 4, 5\} \Rightarrow A \cap B = \{3\} $$

---

### **Difference ($A - B$)**
The difference of two sets $A$ and $B$ contains elements in $A$ but not in $B$.

$$ A - B = \{x \mid x \in A, x \notin B\} $$

**Example:**

$$ A = \{1, 2, 3, 4\}, B = \{3, 4, 5\} \Rightarrow A - B = \{1, 2\} $$

---

## 🏆 Important Laws of Set Theory

| Law | Formula |
|-----|---------|
| **Idempotent Laws** | $ A \cup A = A, \quad A \cap A = A $ |
| **Commutative Laws** | $ A \cup B = B \cup A, \quad A \cap B = B \cap A $ |
| **Associative Laws** | $ (A \cup B) \cup C = A \cup (B \cup C) $ |
| **Distributive Laws** | $ A \cup (B \cap C) = (A \cup B) \cap (A \cup C) $ |
| **De Morgan’s Laws** | $ (A \cup B)^c = A^c \cap B^c $, $ (A \cap B)^c = A^c \cup B^c $ |
