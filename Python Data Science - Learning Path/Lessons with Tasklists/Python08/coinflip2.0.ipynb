Lets simulate the __coin flip__

> I will check the experimental probability of `H` or `T`.

> `Hypothesis` is that the probability will be 50% for the infinite number of tries.

```python
import numpy as np
import matplotlib.pyplot as plt

coin = ['H','T']
rng = np.random.default_rng()
coin_flip = np.logspace(1,3,99, dtype=int)
head = []
for n_rolls in coin_flip:
    flipping = rng.choice(coin,size=n_rolls)
    results = np.unique(flipping,return_counts=True)
    heads = results[1][0]
    head.append(heads)
    print(head)
```

### Here we can observe `np.unique( )` retunrs tuple of 2 arguments (arrays) of which [0] is __values__ and [1] is __counts__.

#### I want to extract the number of `H` received in every flip. 

> To extract the counts of 'H' i wrote the code to choose the second argument inside returned np.unique tuple and that is counts and [0] to choose the first value inside the counts 
which is 'H'. 

> This approach seems to be good in this example but there is a thing:

- `np.unique` sorts the unique values alphabetically, which could be problematic when used in context.

### Lets plot the data with matplotlib.

- x axis will be number in flips in logarithmic space.

- y axis will be `H` received.

- I will also plot the horizontal line where experimental probability is ~50%.

But first lets define theoretical probablity !

```python
plt.subplot()
plt.plot(coin_flip,head, marker='.', color='red', label='Heads')
plt.plot(coin_flip, 0.5*np.array(coin_flip),linestyle="--", color="blue", label="50% Expected")
plt.xscale('log')
plt.yscale('log')
plt.ylabel('Number of H')
plt.xlabel('No. of coin-flips')
plt.legend()
plt.show()
```

### Linear increase of counts `H` with the increase of the coin flips tries.

### What is the experimental probability of receiving H in each flip?

#### Lets see how it increases with coin flips.

```python
coin = ['H','T']
rng = np.random.default_rng()
coin_flip = np.logspace(1.75,7,99, dtype=int)
experminetal_prob_head = []
for n_rolls in coin_flip:
    flipping = rng.choice(coin,size=n_rolls)
    experminetal_prob_head.append(np.mean(flipping=='H'))
```

```python
plt.plot(coin_flip,experminetal_prob_head,marker='.',color='red')
plt.xscale('log')
plt.axhline(0.5,label='experimental probablity')
plt.xlabel('Number of coin flips')
plt.ylabel('Experimental Probability of H')
plt.legend()
plt.grid(alpha=.3)
```
