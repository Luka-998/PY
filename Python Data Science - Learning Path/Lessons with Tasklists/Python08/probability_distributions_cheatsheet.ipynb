## ðŸŽ² Binomial Distribution

**Definition:** The Binomial distribution models the number of successes in `n` independent trials where each trial has a probability `p` of success.

**Formula:**
\[ P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k} \]

**Parameters:**
- `n`: Total number of trials
- `p`: Probability of success in each trial
- `k`: Number of successes


```python
from scipy.stats import binom
import numpy as np
import matplotlib.pyplot as plt

n, p = 10, 0.5  # 10 trials, 50% success probability
x = np.arange(0, n + 1)
pmf = binom.pmf(x, n, p)

plt.bar(x, pmf, color='skyblue')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution')
plt.show()
```

## ðŸŽ¯ Geometric Distribution

**Definition:** The Geometric distribution models the number of trials until the first success occurs.

**Formula:**
\[ P(X = k) = (1 - p)^{k - 1} p \]

**Parameters:**
- `p`: Probability of success in each trial
- `k`: Number of trials until the first success


```python
from scipy.stats import geom

p = 0.5  # Probability of success
x = np.arange(1, 11)
pmf = geom.pmf(x, p)

plt.bar(x, pmf, color='lightcoral')
plt.xlabel('Number of Trials until First Success')
plt.ylabel('Probability')
plt.title('Geometric Distribution')
plt.show()
```

## âš¡ Poisson Distribution

**Definition:** The Poisson distribution models the number of events occurring in a fixed interval when the events happen independently at a constant rate.

**Formula:**
\[ P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!} \]

**Parameters:**
- `Î»`: Average number of occurrences in the interval
- `k`: Number of occurrences


```python
from scipy.stats import poisson

Î» = 4  # Average number of events per interval
x = np.arange(0, 15)
pmf = poisson.pmf(x, Î»)

plt.bar(x, pmf, color='orange')
plt.xlabel('Number of Occurrences')
plt.ylabel('Probability')
plt.title('Poisson Distribution')
plt.show()
```

## ðŸ“ Discrete Uniform Distribution

**Definition:** The Discrete Uniform distribution models a scenario where all outcomes are equally likely over a finite range.

**Formula:**
\[ P(X = k) = \frac{1}{b - a + 1} \]

**Parameters:**
- `a`: Minimum value
- `b`: Maximum value


```python
from scipy.stats import randint

a, b = 1, 6  # Dice roll (1 to 6)
x = np.arange(a, b + 1)
pmf = randint.pmf(x, a, b + 1)

plt.bar(x, pmf, color='green')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Discrete Uniform Distribution (Dice Roll)')
plt.show()
```
