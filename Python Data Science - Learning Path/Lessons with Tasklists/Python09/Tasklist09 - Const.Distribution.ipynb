Exponential distribution is the distribution of time between events in a process in which same events occur continuously and independently at a constant average rate. It's denoted by $\mathcal{E}(\beta)$, where $\beta$ is its parameter, called *scale parameter*, and it denotes an average time by which event should occur.

We can sample from Exponential Distribution using `rng.exponential()`:

so, `scale = beta`

```python
from scipy.stats import expon
```

**06.** Draw a sample of 10000 uniformly distributed random numbers from the Continous Uniform Distribution on [0,1]. Recall the Normal Quantile function: use it to generate 10000 normally distributed random numbers with $\mu=5$ and $\sigma=2$ from the uniformly distributed ones that you have generated. Compute the mean and the standard deviation of the resulting Normal random numbers.

```python
from scipy.stats import uniform
unif_rvs = uniform.rvs(size = 10000)
unif_rvs
```

```python
norm_rvs = norm.ppf(q=unif_rvs, loc=5, scale=2)
norm_rvs_mean = norm_rvs.mean()
norm_rvs_std = norm_rvs.std(ddof=1)
print(f'The mean is {norm_rvs_mean}.')
print(f'The mean is {norm_rvs_std}.')
```
