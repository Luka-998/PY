 ### **1. Line plot**

There are 2 main ways how we can use the matplotlib library. We divide them into explicit and implicit. 

- `Explicitly` creates figures and axes and call methods on them while;

- rely on `py.plot ()` to create `Implicitly` create and manage Figures and Axes  and use pyplot functions for plotting.

#### Objective: **1.1** 

__Objective: Show the trend of sales over months.__

```python
plt.figure(figsize=(15,4))
plt.plot(data_set['Month'], data_set['Sales'], marker='o', color='blue', label='Sales')
plt.title('Trend sales over the period of 12 months')
plt.xlabel('Months')
plt.grid(alpha=.3)
plt.legend()
```

#### Objective: **1.2** 

Compare sales for each month.


```python
plt.figure(figsize=(10,5))
sns.barplot(x='Month', y='Sales', data = data_set, palette='Blues_d')
plt.title('Sales comparison for each month')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
sns.despine()
```

#### Objective: **1.3** 

Visualisation the relationship between adverstising and sales while using the `scatterplot`.


```python
plt.figure(figsize=(12,5))
plt.scatter(data_set['Month'], data_set['Advertising'], label='Relationship', color='red', marker='o')
plt.xlabel('Sales')
plt.title('Advertising')
```

#### Objective: **1.4** 

Visualize the distribution of sales. while using the `histogram`.


```python
plt.figure(figsize=(10,5))
plt.hist(data_set['Sales'], color='skyblue', bins=20)
plt.xlabel('Month')
plt.title('Distribution of sales')
plt.ylabel('Frequency')
plt.xlabel('Sales')
plt.grid(alpha=.2)
plt.show()
```

#### Objective: **1.5** 

Objective: Exercise with `plt.subplots`.

```python
fig, ax = plt.subplots(figsize=(15, 5))
ax.scatter(data_set['Advertising'], data_set['Sales'], color='blue', label='Data Points')
ax.set_title('Advertising vs Sales', fontsize=14)
ax.set_xlabel('Advertising Spend', fontsize=12)
ax.set_ylabel('Sales', fontsize=12)
ax.grid(alpha=0.3)

# Trendline
z = np.polyfit(data_set['Advertising'], data_set['Sales'], 1)  # Linear regression fit
ax.plot(data_set['Advertising'], (data_set['Advertising']), color='red', linestyle='--', label='Trendline')
ax.legend()
plt.show()
```

The scatter plot highlights the relationship between Advertising and Sales, with the trendline showing a positive correlation.
