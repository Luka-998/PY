```python
# .copy() = The .copy() method is used in Python to create a shallow copy of a list, dictionary, or other objects like DataFrames in pandas. 
# A shallow copy means that it creates a new object but does not create copies of the objects inside it. Instead, it references the same objects in memory.

original_list = [1,2,3,4]

a = [1,2,3,4,5]
b=a
b.append(10)
print(a)
print(b)
# So, a = b and b = a in Python mean only that a and b refer to the same object. If you want to assign a to b and continue manipulating b as an independent being, you need to:
a  = [1,2,3,4]
b = a.copy()
a.append(10)
print(b)

# Same is applied to the Dictionaries

a = {'first':1000, 'second':1500}
b = a 
a['third']= 10000
print(a)

a = {'first':100, 'second':1000}
b = a.copy() #b remained the same, with 2 keys and 2 values, while a is appended
a['third'] = 10000
print(b)

#Sets

    #In lists and tuples, order matters.
    #In dictionaries, keys matter while order is unimportant.
    #Sets are like dictionaries without keys, and order is unimportant.

my_set = {0,1,2} # example how order is not important in dictionaries
0 in my_set 

#Little about Set theory!
#1) Union "|" 
#2) intersection "&" (every element that is present in both set A and set B)
#3) set differencies "A - B" (every element that is in the set A but not in the set B)

a = {1,2}
b = {1,3}
a | b # 1 ,2 ,3 
a & b # 1
a - b #2

#check if a is subset of b

a = {1,2,3,4}
b = {7,4,3,2,1,0}
a.issubset(b) #Yes it is "True"
b.issubset(a) # False
```
