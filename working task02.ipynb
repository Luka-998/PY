```python
    #**02.** Using `zip()` function create a data frame containing three rows and five columns. Then, using data frame indexing, extract 2nd, 3rd and 4th column of a middle row. 
name = ['Luka', 'Nevena', 'Matija']
age = [15, 22, 19]
salary = [1200, 1400, 1000]
height = [180, 190, 205]
hair = ['Brown', 'White', 'Orange']

my_dict = pd.DataFrame(list(zip(name, age, salary, height, hair)),
                       columns=['name', 'age', 'salary', 'height', 'hair'],
                       index=['Row1', 'Row2', 'Row3'])
display(my_dict)

display(my_dict.loc['Row2'][1:4])
display(my_dict.loc['Row2',['age','salary','height']])
display(my_dict.iloc[1][1:4])   
```

```python
#**03.** Load `MovieRatings.csv` dataset and save it as a `data_set` variable.

data_set = pd.read_csv('_data\MovieRatings.csv')
display(data_set)

#**04.** From `data_set` create a data frame showing movies with their IMDB and Rotten Tomatoes rating.

data_set = pd.DataFrame(data_set)
display(data_set.loc[:,['IMDB','RottenTomatoes']])

#**05.** `.max()` is a method that you can use to find a maximal value for a given column. 
# For example, `data_set['Metacritic'].max()` will give you the highest Metacritic movie rating. Using similar appropriate methods find:
#    1)- The lowest IMDB rating
#    2)- Average Rotten Tomatoes rating
#    3)- Median Fandango stars rating


data_set['IMDB'].max() #maximal value in the IMDB column 

#1)
data_set['IMDB'].min() #lowestimated in the column

#2)
data_set['RottenTomatoes'].mean() #average

#3)
data_set['Fandango_Stars'].median()
```

```python
# **06.** Plot a histogram of IMDB ratings.

#Could be done like this, with matplotlib library but also as just calling .hist() method, see below:

import matplotlib.pyplot as plt
display(data_set)
plt.hist(data_set['IMDB'], bins=10, color='red', edgecolor='black')
plt.title('Histogram of IMDB ratings')
plt.xlabel('IMDB rating')
plt.ylabel('Number of movies')
plt.show()
```

```python
#**08.** Use `normalize=True` argument in the `.value_counts()` method on `'Fandango_Stars` column to obtain ratios of the previous result. Then plot the new result as a pie chart.
new_res = data_set['Fandango_Stars'].value_counts(normalize=True)
new_res.plot.pie();
```

```python
#**09.** Using an appropriate method sort all the movies according to their IMDB ratings. Then, using a proper argument sort them in descending order. 
data_set = data_set[sorted(data_set.columns)] # This could be done on columns
display(data_set.sort_values(by= 'IMDB', ascending=True))
```

```python

# **10.** Select all movies with Rotten Tomatoes rating of 100.
data_set[data_set['RottenTomatoes'] == 100]

# **11.** Select all movies with Metacritic score lower than 20.
data_set[data_set['Metacritic']<20]

#*12.** Select all movies with IMDB rating of 8 or higher.

data_set[data_set['IMDB']>=8]
```

```python
# **13.** Using the following logical operators you can combine several logiacl conditions to extract the data based on those conditions:   
       # - ~ NOT
       # - | OR
       # - & AND

# For example, `data_set[(cond_1) & (cond_2)]` will extract the data meeting both conditions `cond_1` AND `cond_2`. Using these logical operators extract all movies:

data_set[~(data_set['IMDB']>4.5)] #which are not greater than 4.5 by= IMDB rating

data_set[(data_set['RottenTomatoes']>99) | (data_set['IMDB']==4.4)]

data_set[(data_set['RottenTomatoes']>90) & (data_set['IMDB']>8)]

# **14.** Plot a scatterplot showing the correlation between Metacritic and IMDB scores.

data_set.plot.scatter(x='Metacritic', y='IMDB');

#The scatter plot helps identify relationships or correlations between the two variables. For example:

       #Positive Correlation: If the points form an upward trend, higher Metacritic scores generally correspond to higher IMDB scores.
       #Negative Correlation: If the points form a downward trend, higher Metacritic scores correspond to lower IMDB scores.
       #No Correlation: If the points are scattered randomly, thereâ€™s likely no relationship between the two variables.
#this is likely positive correlation 
```
